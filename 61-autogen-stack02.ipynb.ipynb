```python
import pandas as pd
from pandas_datareader import data as web
import datetime
import matplotlib.pyplot as plt

# Specify the start date
start_date = datetime.datetime(2020, 3, 1)

# Fetch the current date
now = datetime.datetime.now()

# Fetch the stock data
nvda = web.DataReader('NVDA', 'yahoo', start_date, now)
tesla = web.DataReader('TSLA', 'yahoo', start_date, now)

# Calculate the change in stock price
nvda['Change'] = (nvda['Close'] - nvda['Open']) / nvda['Open'] * 100
tesla['Change'] = (tesla['Close'] - tesla['Open']) / tesla['Open'] * 100

# Plot the stock price change
fig, ax = plt.subplots()

ax.plot(nvda.index, nvda['Change'], label='NVIDIA')
ax.plot(tesla.index, tesla['Change'], label='TESLA')

ax.set_xlabel('Date')
ax.set_ylabel('Change (%)')
ax.set_title('Stock Price Change for NVDA and TESLA since 2020-03-01')
ax.legend()

plt.show()
```
